```python
from text_labler import AmharicNERLabeler
```

```python
df = pd.read_csv("../data/tokenized_data.csv")
```

```python
df.head()
```

```python
list(df["cleaned_text"])
```

```python
ner_labeler = AmharicNERLabeler()
```

```python
tokenized_text =   df["cleaned_text"].dropna().tolist()
df = pd.Series(tokenized_text).reset_index(name='message')
```

```python
df['token'] = df['message'].apply(lambda x:x.split())
labeled = ner_labeler.label_dataframe(df,'token')

ner_labeler.save_conll_format(labeled, '../lebeled_conll_data.conll')
```
